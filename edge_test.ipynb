```python
from oraclecit import * 
from itertools import combinations
from tqdm import tqdm 
from math import factorial
```

```python
def comb(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))
def find_min_separating_set(f_dict, x_feature, y_feature, targets, citobject):
    x_idx = f_dict[x_feature]
    y_idx = f_dict[y_feature]
    all_indices = list(f_dict.values())
    target_indices = set(f_dict[f] for f in targets)

    cond_candidates = [i for i in all_indices if i != x_idx and i != y_idx]

    for k in range(len(cond_candidates) + 1):
        k1 = len(cond_candidates) - k  
        # k1 = k 
        total = comb(len(cond_candidates), k1)  
        for subset in tqdm(combinations(cond_candidates, k1), total=total, desc=f"Trying size {k1}"):
            if citobject.query(A=x_idx, B=y_idx, Z=subset):
                cleaned_subset = [i for i in subset if i not in target_indices]
                return cleaned_subset, subset
    return None
```

```python
from collections import defaultdict, deque
```

```python

def topological_sort(nodes, edges):
    # nodes: list of all node labels (e.g., [A, B, C, D])
    # edges: list of directed edges (e.g., [(A, B), (A, C)] means A → B and A → C)

    in_degree = {node: 0 for node in nodes}
    graph = defaultdict(list)

    # Build graph and compute in-degrees
    for u, v in edges:
        graph[u].append(v)
        in_degree[v] += 1

    # Queue for nodes with no incoming edges
    queue = deque([node for node in nodes if in_degree[node] == 0])
    topo_order = []

    while queue:
        node = queue.popleft()
        topo_order.append(node)

        for neighbor in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    # Check for cycle (i.e., not all nodes were included)
    if len(topo_order) != len(nodes):
        raise ValueError("Graph contains a cycle, topological sort not possible.")

    return topo_order
```

```python

def find_cycle(nodes, edges):
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)

    visited = set()
    rec_stack = set()
    parent = {}

    def dfs(node):
        visited.add(node)
        rec_stack.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                parent[neighbor] = node
                cycle = dfs(neighbor)
                if cycle:
                    return cycle
            elif neighbor in rec_stack:
                # Cycle found: reconstruct path
                cycle = [neighbor]
                current = node
                while current != neighbor:
                    cycle.append(current)
                    current = parent[current]
                cycle.append(neighbor)
                cycle.reverse()
                return cycle
        rec_stack.remove(node)
        return None

    for node in nodes:
        if node not in visited:
            parent[node] = None
            cycle = dfs(node)
            if cycle:
                return cycle  # Return the first detected cycle

    return None  # No cycle found
```
